```python
import pandas as pd

df = pd.read_csv('train.csv', low_memory=False)
df
```

### Evaluating and unpacking 'device' column with json values to create seperate columns

```python
# Replacing 'true/false' with 'True/False' so eval does not throw an error

df['device'] = df['device'].str.replace('false', 'False')
df['device'] = df['device'].str.replace('true', 'True')

# Evaluate JSON string to a JSON object
df['device'] = df['device'].map(eval)

# Create seperate columns for each JSON key
df = pd.concat([df.drop(['device'], axis=1), df['device'].apply(pd.Series)], axis=1)
```

### Seperating target and features to filter out columns on the basis of null values and unique values

```python
y = df['transactionRevenue']
y.notnull().sum()
```

```python
y = y.fillna(0.0)
```

```python
X = df.drop(['transactionRevenue'], axis=1)
```

```python
# filtering out columns that have less than 90% non-NaN values

X = X.dropna(axis=1, thresh=int(0.9*len(X)))
X.info()
```

```python
# Setting the null values to median in pageviews column

pageviews_median = X['pageviews'].median()
X['pageviews'] = X['pageviews'].fillna(pageviews_median)
```

```python
# print and drop columns with single unique value

columns = X.columns.tolist()

for col in X.columns:
    if len(X[col].unique()) == 1:
        print(col)
        print(X[col].unique())
        print('\n')
        
        X = X.drop(col, axis=1)
        
X.info()
```

```python
# Dropping Id columns

X = X.drop(['fullVisitorId', 'sessionId', 'visitId'], axis=1)
```

```python
# Printing unique values in each column

for col in X.columns:
    print(col, '\n')
    print(len(X[col].unique()))
    print(X[col].unique())
    print('\n')
```

### Creating day, month and year columns

```python
# Seperating 'day', 'month' and 'year' and dropping 'date' column

X['date'] = X['date'].astype('|S')

X['year'] = X['date'].str.slice(0, 4).astype(int)
X['month'] = X['date'].str.slice(4, 6).astype(int)
X['day'] = X['date'].str.slice(6, 8).astype(int)

X.drop('date', axis=1, inplace=True)

print(sorted(X['year'].unique()))
print(sorted(X['month'].unique()))
print(sorted(X['day'].unique()))
```

```python
from datetime import datetime

X['hour'] = (X['visitStartTime'].apply(lambda x: str(datetime.fromtimestamp(x).hour))).astype(int)

X.drop(['visitStartTime'], axis=1, inplace=True)

print(sorted(X['hour'].unique()))
```

```python
# Creating sin and cos features considering the cyclic nature of months

import numpy as np

X['month_sin'] = np.sin(X['month'] * 2. * np.pi / len(X['month'].unique()))
X['month_cos'] = np.cos(X['month'] * 2. * np.pi / len(X['month'].unique()))

X['day_sin'] = np.sin(X['day'] * 2. * np.pi / len(X['day'].unique()))
X['day_cos'] = np.cos(X['day'] * 2. * np.pi / len(X['day'].unique()))

X['hour_sin'] = np.sin(X['hour'] * 2. * np.pi / len(X['hour'].unique()))
X['hour_cos'] = np.cos(X['hour'] * 2. * np.pi / len(X['hour'].unique()))
```

```python
# concatenating features and target columns to save to a file

unpacked_filtered_df = pd.concat([X, y], axis=1)
```

```python
# Save the unpacked and filtered data

unpacked_filtered_df.to_csv('unpacked_filtered_data.csv', index=False)
```
